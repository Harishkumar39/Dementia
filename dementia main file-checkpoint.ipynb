```python
def univariate_percent_plot(cat):
    fig = plt.figure(figsize=(18,12))
    cmap=plt.cm.Blues
    cmap1=plt.cm.coolwarm_r
    ax1 = fig.add_subplot(221)
    ax2 = fig.add_subplot(222)
    
    result = df_dementia1.groupby(cat).apply (lambda group: (group.CDR == 'Normal').sum() / float(group.CDR.count())).to_frame('Normal')
    result['Dementia'] = 1 -result.Normal
    result.plot(kind='bar', stacked = True,colormap=cmap1, ax=ax1, grid=True)
    ax1.set_title('stacked Bar Plot of '+ cat +' (in %)', fontsize=14)
    ax1.set_ylabel('% Dementia status (Normal vs Dementia)')
    ax1.legend(loc="lower right")
    group_by_stat = df_dementia1.groupby([cat, 'CDR']).size()
    group_by_stat.unstack().plot(kind='bar', stacked=True,ax=ax2,grid=True)
    ax2.set_title('stacked Bar Plot of '+ cat +' (in %)', fontsize=14)
    ax2.set_ylabel('Number of Cases')
    plt.show()



def cat_CDR(n):
    if n == 0:
        return 'Normal'
    
    else:                                         
        return 'Dementia'

```

```python
def cat_MMSE(n):
    if n >= 24:
        return 'Normal'
    elif n <= 9:
        return 'Severe'
    elif n >= 10 and n <= 18:
        return 'Moderate'
    elif n >= 19 and n <= 23:                                   
        return 'Mild'

```

```python
df_dementia1['MMSE'] = df_dementia1['MMSE'].apply(lambda x: cat_MMSE(x))
```

# Building model 

```python
df["Group"].value_counts()
```

<b>The dataset is balanced</b>

```python
# X=df.drop(["Group"],axis=1)
# y=df["Group"]
# Modified
df.drop(["Group"], axis=1, inplace = True)

X=df.drop(["CDR"],axis=1)
y=df["CDR"]
```

```python
#X=sm.add_constant(X)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_train
```

### <span style="color:Red;font-size:1.4em;"><b>KNN Algorithm</b></span> 

```python
knn_classification = KNeighborsClassifier(n_neighbors = 3)
knn_model = knn_classification.fit(X_train, y_train)

Y_pred_knn = knn_model.predict(X_test)

score_knn = round(accuracy_score(Y_pred_knn,y_test)*100,2)

print("The accuracy score achieved using KNN is: "+str(score_knn)+" %")
```

### <span style="color:Red;font-size:1.4em;"><b>Ridge Classifier</b></span> 

```python
from sklearn.linear_model import RidgeClassifier
from sklearn.metrics import accuracy_score

# Create a Ridge Classifier model
ridge_classifier = RidgeClassifier(alpha=1.0)

# Fit the model on your training data
ridge_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred_ridge = ridge_classifier.predict(X_test)

# Calculate the accuracy score
accuracy_r = accuracy_score(y_test, y_pred_ridge)
print("Accuracy achieved using Ridge Classifier: {:.2f}%".format(accuracy_r * 100))
```

### <span style="color:Red;font-size:1.4em;"><b>Decision Tree Classifier</b></span> 

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train) 
ypred_dt = model.predict(X_test) 
score_dt = round(accuracy_score(ypred_dt,y_test)*100,2)

print("The accuracy score achieved using Decision Tree is: "+str(score_dt)+" %")
```

### <span style="color:Red;font-size:1.4em;"><b>CONVOLUTION NEURAL NETWORK</b></span> 

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout,Flatten
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_reshaped = X_train_scaled.reshape((X_train_scaled.shape[0], X_train_scaled.shape[1], 1))
X_test_reshaped = X_test_scaled.reshape((X_test_scaled.shape[0], X_test_scaled.shape[1], 1))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, MaxPooling1D, Dropout, Flatten, Dense

cnn_model = Sequential()
cnn_model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2]), padding='same'))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Dropout(0.2))
cnn_model.add(Conv1D(filters=32, kernel_size=2, activation='relu', padding='same'))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Dropout(0.2))
cnn_model.add(Flatten())
cnn_model.add(Dense(units=1, activation='sigmoid'))
cnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history_cnn = cnn_model.fit(X_train_reshaped, y_train, epochs=10, batch_size=32, validation_data=(X_test_reshaped, y_test))

```

### <span style="color:Red;font-size:1.4em;"><b>LSTM ALGORITHM</b></span> 

```python
model = Sequential()
model.add(LSTM(64, input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2])))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train_reshaped, y_train, epochs=10, batch_size=32, validation_data=(X_test_reshaped, y_test))
```

```python
validation_accuracy_CNN = history_cnn.history['val_accuracy'][-1]
print("Validation accuracy achieved using the CNN model is: {:.2f}%".format(validation_accuracy_CNN * 100))
```

```python
validation_accuracy = history.history['val_accuracy'][-1]
print("Validation accuracy achieved using the LSTM model is: {:.2f}%".format(validation_accuracy * 100))
```

```python
print("The accuracy score achieved using KNN is: "+str(score_knn)+" %")
print("Accuracy achieved using Ridge Classifier: {:.2f}%".format(accuracy_r * 100))
print("The accuracy score achieved using Decision Tree is: "+str(score_dt)+" %")
print("Validation accuracy achieved using the CNN model is: {:.2f}%".format(validation_accuracy_CNN * 100))
print("Validation accuracy achieved using the LSTM model is: {:.2f}%".format(validation_accuracy * 100))
```
